# 11 Write a Python program to generate the next 15 leap years starting from a given year. Populate the leap years into a list and display the list.

```python
def leap_year(yr):
  if ((yr  % 400 == 0) or (yr% 4 == 0 and yr % 100 != 0)):
    return True

  else:
    return False

year = int(input("Enter a year:"))
cnt = 0
leap_year_list = []

while cnt < 15:
  if leap_year(year):
    print(f'{year} is a leap year\n')
    leap_year_list.append(year)
    cnt+=1
  year+=1


print(leap_year_list)
```

# 21.
 Care hospital wants to know the medical speciality visited by the maximum number of patients. Assume that the patient id of the patient along with the medical speciality visited by the patient is stored in a list. The details of the medical specialities are stored in a dictionary as follows:

* {
"P":"Pediatrics",
"O":"Orthopedics",
"E":"ENT
}


* Write a function to find the medical speciality visited by the maximum number of patients and return the name of the speciality.


* Note:
* 1. Assume that there is always only one medical speciality which is visited by maximum number of patients.
* 2. Perform case sensitive string comparison wherever necessary.


```python
def most_visited_speciality(patient_visits, specialty_dict):
    # Create a dictionary to count the number of visits to each specialty
    visit_count = {}

    # Iterate through the list of patient visits
    for patient_id, specialty_code in patient_visits:
        # Increment the count for the specialty code
        if specialty_code in visit_count:
            visit_count[specialty_code] += 1
        else:
            visit_count[specialty_code] = 1

    # Find the specialty code with the maximum count
    max_count = -1
    max_specialty_code = None

    for code, count in visit_count.items():
        if count > max_count:
            max_count = count
            max_specialty_code = code

    # Return the full name of the specialty with the maximum count
    return specialty_dict[max_specialty_code]






# Example usage:
patient_visits = [
    ('P001', 'P'),
    ('P002', 'E'),
    ('P003', 'P'),
    ('P004', 'O'),
    ('P005', 'P'),
    ('P006', 'E')
]

specialty_dict = {
    "P": "Pediatrics",
    "O": "Orthopedics",
    "E": "ENT"
}

print(most_visited_speciality(patient_visits, specialty_dict))
```
