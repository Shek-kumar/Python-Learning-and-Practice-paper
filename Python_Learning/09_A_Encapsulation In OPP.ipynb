```python
class ATM:

    # constructor( __init__ ) --> here we declare all the variables inside the method __init__ called constructor

    def __init__(self):
        print("Welcome to login Page !!")
        self.pin = ""
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input('''
            Hello! How would you like to proceed?
            1. Enter 1 to create your pin
            2. Enter 2 to Deposit money
            3. Enter 3 to withdraw money
            4. Enter 4 to Check your balance
            5. Enter 5 to Exit
        ''')

        if user_input == '1':
            print("Create your pin")
            self.create_pin()

        elif user_input == "2":
            print("Deposit Money")
            self.deposit()

        elif user_input == "3":
            print("Withdraw Money")
            self.withdraw()

        elif user_input == '4':
            print("Checking Balance")
            self.check_bal()

        else:
            print("Thank You Visit Again")

    def create_pin(self):
        self.pin = input("Enter your Pin:")
        print("Pin Created Successfully")

    def deposit(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            amount = int(input("Enter the Amount of money you want to deposit: "))
            self.balance += amount
            print("Money Deposited Successfully")
        else:
            print("Incorrect Pin !!")

    def withdraw(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            amount = int(input("Enter the Amount of money you want to withdraw: "))

            if self.balance >= amount:
                self.balance -= amount
                print("Money Withdrawn Successfully")
            else:
                print("Insufficient Balance !!")
        else:
            print("Incorrect Pin !!")

    def check_bal(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            print("Your Balance is:", self.balance)
        else:
            print("Incorrect Pin !!")


```

# **Collection of Objects**

```python
class Customer:

  def __init__(self,name,age,gender):
    self.name=name
    self.age=age
    self.gender=gender

  def intro(self):
    print("Hello I am {} . I am {} years old".format(self.name,self.age))

c1=Customer("Sam",20,"Male")
c2=Customer("Anika",19,"Female")
c3=Customer("Shekhar",21,"Male")

L=[c1,c2,c3]

for i in L:
  i.intro()
  print(i.gender)
```

# Using Static Methods

```python
class Atm:

  # Static varaibles
  __counter=1


  def __init__(self):
    self.pin=""
    self.balance=0
    self.serial =  Atm.__counter
    Atm.__counter+=1

  @staticmethod
  def get_counter():
    return Atm.__counter


  @staticmethod
  def set_counter(x):
    if type(x)==int:
      Atm.__counter=x

    else:
      print("Not allowed !")



```

```python
Atm.get_counter()
```

```python
Atm.set_counter(20)
```
