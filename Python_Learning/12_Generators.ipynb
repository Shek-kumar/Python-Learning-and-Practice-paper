#Example 2

```python
def sq(num):
  for i in range(1,num+1):
    yield i**2
```

```python
gen=sq(9)

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))

for i in gen:
  print(i)
```

# Range in Generator

```python
def mera_range(start,end):
  for i in range(start,end):
    yield i
```

```python
GEN=mera_range(15,30)

print(next(GEN))
print(next(GEN))
print(next(GEN))
print(next(GEN))
print(next(GEN))

for i in GEN:
  print(i)
```

# Generator Expression similar to list comprehension

```python
# list comprehension

L=[i**2 for i in range(1,10)]
print(L)
```

```python
# generator expression

L=(i**2 for i in range(1,10))
print(L)
```

3. Representing Infinite Streams

```python
def all_even():
  n=0
  while True:
    yield n
    n+=2
```

```python
even_num_gen=all_even()

next(even_num_gen)
```

```python
next(even_num_gen)
```

4. Chaining Generators

```python
def fibonaci(nums):
  a=0
  b=1
  for i in range(nums):
    a,b=b,a+b
    yield a

def square(nums):
  for i in nums:
    yield i**2


print(sum(square(fibonaci(10))))
```
