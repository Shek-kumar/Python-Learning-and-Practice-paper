#**Creating a class for Atm machine**

```python
class ATM:

    # constructor( __init__ ) --> here we declare all the variables inside the method __init__ called constructor

    def __init__(self):
        print("Welcome to login Page !!")
        self.pin = ""
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input('''
            Hello! How would you like to proceed?
            1. Enter 1 to create your pin
            2. Enter 2 to Deposit money
            3. Enter 3 to withdraw money
            4. Enter 4 to Check your balance
            5. Enter 5 to Exit
        ''')

        if user_input == '1':
            print("Create your pin")
            self.create_pin()

        elif user_input == "2":
            print("Deposit Money")
            self.deposit()

        elif user_input == "3":
            print("Withdraw Money")
            self.withdraw()

        elif user_input == '4':
            print("Checking Balance")
            self.check_bal()

        else:
            print("Thank You Visit Again")

    def create_pin(self):
        self.pin = input("Enter your Pin:")
        print("Pin Created Successfully")

    def deposit(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            amount = int(input("Enter the Amount of money you want to deposit: "))
            self.balance += amount
            print("Money Deposited Successfully")
        else:
            print("Incorrect Pin !!")

    def withdraw(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            amount = int(input("Enter the Amount of money you want to withdraw: "))

            if self.balance >= amount:
                self.balance -= amount
                print("Money Withdrawn Successfully")
            else:
                print("Insufficient Balance !!")
        else:
            print("Incorrect Pin !!")

    def check_bal(self):
        passcode = input("Input Your pin:")

        if passcode == self.pin:
            print("Your Balance is:", self.balance)
        else:
            print("Incorrect Pin !!")


```

#**Creating a Fraction Dattype along with Addition**

```python
import math as m

class Fraction:

  def __init__(self,numerator,denominator):
    self.numerator=numerator
    self.denominator= denominator

  def __str__(self):
    print("Welcome to Fraction datatype !!")
    return('{}/{}'.format(self.numerator,self.denominator))


  def simplify(self):
    temp=m.gcd(self.numerator,self.denominator)
    self.numerator=self.numerator//temp
    self.denominator=self.denominator//temp

  def __add__(self,other):
    new_num=self.numerator*other.denominator+self.denominator*other.numerator
    new_denominator=self.denominator*other.denominator
    res_frac=Fraction(new_num,new_denominator)
    res_frac.simplify()

    return res_frac
```

```python
x=Fraction(4,6)
y=Fraction(6,4)

print(x)
print(y)
print(x+y)
```

#**Creating a Fraction Dattype along with Subtraction**

```python
import math as m

class Fraction:

  def __init__(self,numerator,denominator):
    self.numerator=numerator
    self.denominator= denominator

  def __str__(self):
    return('{}/{}'.format(self.numerator,self.denominator))


  def simplify(self):
    temp=m.gcd(self.numerator,self.denominator)
    self.numerator=self.numerator//temp
    self.denominator=self.denominator//temp

  def __add__(self,other):
    new_num=self.numerator*other.denominator+self.denominator*other.numerator
    new_denominator=self.denominator*other.denominator
    print("Without Simplication {}/{}".format(new_num,new_denominator))

    res_frac=Fraction(new_num,new_denominator)
    res_frac.simplify()
    return res_frac


  def __sub__(self,other):
      new_num=self.numerator*other.denominator-self.denominator*other.numerator
      new_denominator=self.denominator*other.denominator
      print("Without Simplication {}/{}".format(new_num,new_denominator))

      res_frac=Fraction(new_num,new_denominator)
      res_frac.simplify()
      return res_frac


  def __mul__(self,other):
      new_num=self.numerator*other.numerator
      new_denominator=self.denominator*other.denominator
      print("Without Simplication {}/{}".format(new_num,new_denominator))

      res_frac=Fraction(new_num,new_denominator)
      res_frac.simplify()
      return res_frac


  def __truediv__(self,other):
      new_num=self.numerator*other.denominator
      new_denominator=self.denominator*other.numerator
      print("Without Simplication {}/{}".format(new_num,new_denominator))

      res_frac=Fraction(new_num,new_denominator)
      res_frac.simplify()
      return res_frac
```

```python
x=Fraction(4,6)
y=Fraction(6,4)

print(x)
print(y)
print(x+y)

print(x-y)
print(x*y)
print(x/y)
```
