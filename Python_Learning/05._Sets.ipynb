* **Sets don't allow duplicates**
* **No indexing/slicing is there**
* **don't allow a mutable datatypes inside it**
* **set itself is mutable datatype**

# creating a set

```python
# create a empty set

s={}
print(type(s))

s1=set()
print("Empty set is created by using set function only")
print(type(s1))
```

```python
s2={1,2,3,4,5}
print(s2)
print(type(s2))
```

```python
s3={1,20.3,'c'}
print(s3)
print(type(s3))
```

```python
# Duplicates

s4={1,1,2,3,4,5,6,6}
print(s4)
type(s4)
```

```python
s5={(1,2,3,4),"Hi"}
print(s5)
type(s5)
```

```python
s6={1,2,3,2,"Hi"}
print(s6)
type(s6)
```
