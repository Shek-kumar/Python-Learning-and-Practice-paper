```python
l1=[1,2,3,4,5,4,3]

l2=[]

for i in l1:
  if i  not in l2:
    l2.append(i)
```

```python
print(l2)
```

```python
l1=[1,2,3,4,5,4,3]

l2=[]

for i in l1:
  if i not in (l1[l1.index(i)+1:] or l1[:l1.index(i)]):
    l2.append(i)

print(l2)
```
