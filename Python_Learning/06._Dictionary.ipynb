#**Dictionary** --> **key value based inside a curly braces separated by comma**

* **Dictionary has no indexing**
* **It is mutable**
* **keys-->immutable but values can be mutable**
* **keys should be unique**

```python
# create a dictionary

cap = {"India": "New Delhi", "Pakistan": "Islamabad", "Bangladesh": "Dhaka"}
print(cap)
```

```python
type(cap)
```

```python
# no indexing
cap[0]
```

```python
# key can't be mutable

d1={[1,2,3]:"Shek","age":19}
d1
```

```python
d1={{1,2,3}:"Shek","age":19}

d1
```

```python
d1={(1,2,3):"Shek","age":19}

d1
```

```python
# key should be unique

d2={"name":"Shek","name":"Sheh"}

d2
```
