#Factorial

```python
# using recursion

def fac(n):
  if n==0 or n==1:
    return(1)

  else:
    return(n*fac(n-1))
```

```python
fac(5)
```

#palindrome

```python
# using for loop

def palin_loop(str):
  L=list(str)
  reverse_list=L[::-1]

  for i in range(len(L)):
    if L[i]==reverse_list[i]:
      continue
    else:
      return False
  return True
```

```python
palin_loop("malayalam")
```

```python
palin_loop("jelicdkwj")
```

```python
# USING RECURSION

def palindrome(S):
  if(len(S)<=1):
    return True

  elif S[0]==S[-1]:
    return palindrome(S[1:-1])

  else:
    return False
```

```python
palindrome("madam")
```

```python
palindrome("Shek")
```

```python
palindrome("malayalam")
```

```python
palindrome("abba")
```

```python
import time

def fib(m):
  if m==0 or m==1:
    return(1)

  else:
    return(fib(m-1)+fib(m-2))

strt=time.time()
print(fib(12))
print(time.time()-strt)
```

#We can solve this expomnential time by **Memoization(Dynamic  programming)** where we store the values once calculated and use that vaue when required again rather than computing that value again and again

```python
def memo_fib(m,d):

  if m in d:
    return(d[m])

  else:
    d[m]=(memo_fib(m-1,d)+memo_fib(m-2,d))
    return(d[m])

d={0:1,1:1}
strt=time.time()
print(memo_fib(12,d))
print(time.time()-strt)
```

```python
strt=time.time()
print(memo_fib(40,d))
print(time.time()-strt)
```

```python
strt=time.time()
print(memo_fib(512,d))
print(time.time()-strt)
```

```python
def pow_set(*L1):
  L=list(L1)
  res=[]
  if len(L)==0:
    return res
  elif len(L)==1:
    return([res,L])

  else:
    return(pow_set(*L[1:])+[x+[L[0]] for x in pow_set(*L[1:])])
```

#**Higher Order Function(HOF)**

**these are the functions for which input is either a function or return is either a function**

**program for getting sum of even,odd and divisible by 3 numbers of a list using higher order functions**

```python
def sum(func,L):
  sum=0
  for i in L:
    if(func(i)):
      sum+=i
  return sum
```

```python
L=[25,36,2,41,45,6,9,89,7,88]

# lambda function which will be input for hof
a= lambda x:x%2==0
b= lambda x:x%2!=0
c=lambda x: x%3==0

sum(a,L)
```

```python
sum(b,L)
```

```python
sum(c,L)
```
