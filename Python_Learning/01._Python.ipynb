```python
N=int(input("Enter the number for which you want to print the table:"))

i=0
while(i<10):
  print(N,'*',i+1,'=',N*(i+1))
  i=i+1
```

#**Guessing Game**

```python
import random
n=random.randint(0,100)

guess=int(input("Enter your first guess:"))
cnt=1
while(guess!=n):
  if(guess>n):
    print("Your guess is wrong and please select a smaller number:")
    guess=int(input("Enter your guess again:"))
    cnt=cnt+1

  elif(guess<n):
    print("Your guess is wrong and please select a larger number:")
    guess=int(input("Enter your guess again:"))
    cnt=cnt+1

print("Eureka correctly guessed!!")
print("You guessed the number correctly in ",cnt ,"attempts")
```
