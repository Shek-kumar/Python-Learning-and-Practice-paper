# Changing Profile of a customer using method of other class

```python
class Customer:

  def __init__(self,name,age,gender,address):
    self.name=name
    self.age=age
    self.gender=gender
    self.address=address

  def edit_profile(self,new_name,new_city,new_pin,new_state):
    self.name=new_name
    self.address.change_address(new_city,new_pin,new_state)     # using method of another class


class Address:

  def __init__(self,city,pincode,state):
    self.city=city
    self.pincode=pincode
    self.state=state


  def change_address(self,new_city,new_pincode,new_state):
    self.city=new_city
    self.pincode=new_pincode
    self.state=new_state

```

```python
add =Address("Gaya",823001," Bihar")

c1=Customer("James",22,"Male",add)
print(c1.name)
print(c1.age)
print(c1.gender)
print(c1.address.city)
print(c1.address.pincode)
print(c1.address.state)


c1.edit_profile("Shekhar","Patna",840001,"Bihar")
print()
print("After Changing")
print(c1.name)

print(c1.address.city)
print(c1.address.pincode)
print(c1.address.state)
```

# **Inheritance**

```python
class User:

  def login(self):
    print("login")


  def register(self):
    print("Register")


class Student(User):

  def enroll(self):
    print("Enroll")


  def Query(self):
    print("Ask Your Questions")


class Admin(User):

  def create_course(self):
    print("Create Course")


  def answer(self):
    print("Answer of the query")
```

#Using Super() keyword to call **parent class** methods


```python
class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("Buying a Phone")



class Smartphone(Phone):

  def buy(self):
    print("Buying a Smartphone")
    super().buy()

m=Smartphone(10000,"Black","Apple")

print(m.brand)
print(m.color)
print(m.price)
m.buy()
```

#Using Super() keyword to call **parent class** constructor


```python
class Phone:

  def __init__(self,price,color,brand):
    print("Secondly constructor of the parent class is called !! ")

    self.price=price
    self.color=color
    self.brand=brand



class Smartphone(Phone):

  def __init__(self,price,color,brand,os,ram):
    print("Firstly inside the constructor of the child class is called !! ")
    super().__init__(price,color,brand)
    self.os=os
    self.ram=ram
    print("Finally Inside the SmartPhone Class ")
```

```python
s=Smartphone(10000,"Blue","Nokia","Android",8)
print(s.os)
print(s.ram)
print(s.brand)
print(s.color)
print(s.price)
```

# Example 2 where parent constructor is called

```python
class Parent:

  def __init__(self,num):
    self.__num=num


  def get_num(self):
    return self.__num


class Child(Parent):

  def __init__(self,val,num):
    super().__init__(num)
    self.__val=val


  def get_val(self):
    return self.__val


c=Child(100,10)
print(c.get_num())
c.get_val()
```

#Types of Inheritance

1. **Single level Inheritance**

```python
class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("Buying a Phone")



class Smartphone(Phone):
  pass


m=Smartphone(10000,"Black","Apple")

print(m.brand)
print(m.color)
print(m.price)
m.buy()
```

2. **Multi level Inheritance**

```python
# Grand Parent

class Product:

  def review(self):
    print("Grandparent")
    print("Customer review for the Product ")


# Parent

class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("parent")
    print("Buying a Phone")


  def return_product(self):
    print("Return the product")


# child class
class Smartphone(Product,Phone):
  pass


m=Smartphone(10000,"Black","Apple")


# from grandparent  class
m.review()

# from parent class
m.buy()
```

3. **Hierachial Inheritance**

```python
# Parent class

class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("parent")
    print("Buying a Phone")


  def return_product(self):
    print("Return the product")



# child 1

class SmartPhone(Phone):
  pass



# child 2

class FeaturePhone(Phone):
  pass


s=SmartPhone(10000,'blue',"samsung")

f=FeaturePhone(5000,"black",'Nokia')

print(s.brand)
s.buy()
```

```python
print(f.brand)
f.buy()
```

4. **Multiple Inheritance**

```python
# Parent 1

class Product:

  def review(self):
    print("Parent 1 ")
    print("Customer review for the Product ")


# Parent 2

class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("parent 2")
    print("Buying a Phone")


  def return_product(self):
    print("Return the product")


# child class
class Smartphone(Product,Phone):
  pass
```

```python
m=Smartphone(10000,"Black","Apple")


# from parent 1 class
m.review()

# from parent 2 class
m.buy()
```

**In case if child class has no constructor then constructtor of first parent class will be invoked if there is no constructor then second parent class will constructor will be invoked and so on .**

```python
# Parent 1

class Product:

  def review(self):
    print("Parent 1 ")
    print("Customer review for the Product ")


  def buy(self):
    print("parent 1(Product class)")
    print("Buying a Phone")


# Parent 2

class Phone:

  def __init__(self,price,color,brand):
    self.price=price
    self.color=color
    self.brand=brand

  def buy(self):
    print("parent 2(Phone class)")
    print("Buying a Phone")


  def return_product(self):
    print("Return the product")


```

# **Some examples of Inheritance**

```python
class A:

  def m1(self):
    return 20


class B(A):
  def m1(self):
    return 30

  def m2(self):
    return 40

class C(B):

  def m2(self):
    return 60
```

```python
obj_1=A()
obj_2=B()
obj_3=C()


print(obj_1.m1())
print(obj_2.m1())
print(obj_3.m1())
print(obj_3.m2())
```

```python
class A:

  def m1(self):
    return 20


class B(A):
  def m1(self):
    val=super().m1()+30
    return val


class C(B):

  def m1(self):
    val=self.m1()+20
    return val
```

```python
obj=C()
print(obj.m1())     # will lead to a loop and infinite recursion as self.m1 is inside m1 of class c
```
