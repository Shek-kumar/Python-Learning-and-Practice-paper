#What is an Iteration
Iteration is a general term for taking each item of something, one after another. Any time you use a loop, explicit or implicit, to go over a group of items, that is iteration.

```python
# Example
num = [1,2,3]

for i in num:
    print(i)
```

#What is Iterator
An Iterator is an object that allows the programmer to traverse through a sequence of data without having to store the entire data in the memory

```python

# Example
L = [x for x in range(1,10000)]

#for i in L:
    #print(i*2)

import sys

print(sys.getsizeof(L)/64)

x = range(1,10000000000)

#for i in x:
    #print(i*2)

print(sys.getsizeof(x)/64)
```

#What is Iterable
Iterable is an object, which one can iterate over

It generates an Iterator when passed to iter() method.

```python
# Example

L = [1,2,3]
print(type(L))
# L is an iterable



type(iter(L))
# iter(L) --> iterator
```

#dir(item) -->  method to detect whether a given item is iterable or not

```python
a=23

'__iter__' in (dir(a))
```

# Making our own for loop

```python
def own_loop(iterable):
  iteration = iter(iterable)

  while True:
    try:
      print(next(iteration))
    except StopIteration:
      break
```

```python
a=[1,23,3,4,568,54,65]
(own_loop(a))
```

```python
b=range(1,11)
(own_loop(b))
```

```python
c=(1,2,3,4,5,6)
(own_loop(c))
```

```python
s={45,6,5}
(own_loop(s))
```

```python
d={"India":"Delhi","China":"Beijing","Japan":"Tokyo"}
own_loop(d)
```

# Let's create our own range() function


```python
class mera_range:
  def __init__(self,start=0,end=1):
    self.start=start
    self.end=end

  def __iter__(self):
    return mera_range_iterator(self)


class mera_range_iterator:
  def __init__(self,iterable_obj):
    self.iterable_obj=iterable_obj


  def __iter__(self):
    return self

  def __next__(self):
    if self.iterable_obj.start >= self.iterable_obj.end:
      raise StopIteration

    else:
      temp=self.iterable_obj.start
      self.iterable_obj.start+=1
      return temp
```

```python
x=mera_range(1,10)

for i in x:
  print(i)
```
